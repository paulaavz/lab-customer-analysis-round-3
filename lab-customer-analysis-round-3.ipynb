```python
import pandas as pd
import numpy as np

#1. Show DataFrame info.
data = pd.read_csv('files_for_lab/csv_files/marketing_customer_analysis.csv')
display(data)
```

```python
#2. Describe DataFrame.

data.describe()
#data.describe().T
#data.describe(include='all').T
```

```python
#3. Show a plot of the total number of responses.
#import matplotlib.pyplot as plt
#import seaborn as sns 

data['Response'].value_counts().plot(kind='bar')
```

```python
#4. Show a plot of the response rate by the sales channel.

import matplotlib.pyplot as plt
import seaborn as sns 

# Filtered the DataFrame to include only 'Yes' and 'No' responses
filtered_data = data[data['Response'].isin(['Yes', 'No'])]

# a. Calculated response rates by sales channel
response_rate = filtered_data.groupby('Sales Channel')['Response'].value_counts(normalize=True).unstack()
# Plot the response rates
response_rate.plot(kind='bar', stacked=True)

plt.title('Response Rate by Sales Channel')
plt.xlabel('Sales Channel')
plt.ylabel('Rate')
plt.show()

# b. Calculated response rates by sales channel
#filtered data.groupby (['Sales Channel', 'Response']).size() _unstack() .plot (kind='bar')
# info: size() gets the number of index entries, like count_values
#plt.show()
```

```python
#5. Show a plot of the response rate by the total claim amount.

#import matplotlib.pyplot as plt
#import seaborn as sns 

sns.barplot(x='Response', y='Total Claim Amount', data=data)
plt.show()
```

```python
#6. Show a plot of the response rate by income.

#import matplotlib.pyplot as plt
#import seaborn as sns 

sns.barplot(x='Response', y='Income', data=data)
plt.show()
```
